```python
import pandas as pd
import matplotlib.pyplot as plt
```

#TURMA HUMANAS

```python
turma = pd.read_excel('/content/mydrive/MyDrive/semestre2.xlsx')
turma
```

#CARREIRA

```python
import matplotlib.pyplot as plt

# Supondo que turma seja um DataFrame do pandas e 'CARREIRA' uma coluna nele
profissao_counts = turma['CARREIRA'].value_counts()

plt.figure(figsize=(5, 3))
bars = plt.bar(profissao_counts.index, profissao_counts.values, color=['blue', 'green', 'red'])
plt.xlabel('Profissão')
plt.ylabel('Quantidade de Alunos')
plt.title('Quantidade de Alunos por Profissão')
plt.xticks(rotation=45, ha='right')  # Ajusta a rotação e a posição horizontal

plt.show()
```

#AREA

```python
# Criar o gráfico de barras para o total de alunos por profissão
profissao_counts = turma['AREA'].value_counts()
plt.figure(figsize=(5, 3))
profissao_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel('Profissão')
plt.ylabel('Quantidade de Alunos')
plt.title('Quantidade de Alunos por Profissão')
plt.xticks(rotation=0)
plt.show()
```

#PREFERENCIA UNIVERSIDADE

```python
 #Criar o gráfico de barras para o total de alunos por profissão
profissao_counts = turma['PREFERENCIA UNIVERSIDADE'].value_counts()
plt.figure(figsize=(5, 3))
profissao_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel('Profissão')
plt.ylabel('Quantidade de Alunos')
plt.title('PREFERENCIA UNIVERSIDADE')
plt.xticks(rotation=0)
plt.show()
```

#EXCEL CONHECIMENTO

```python
# Criar o gráfico de barras para o uso de ChatGPT
chatgpt_counts = turma['EXCEL CONHECIMENTO'].value_counts()
plt.figure(figsize=(5, 3))
chatgpt_counts.plot(kind='bar', color=['green', 'blue','yellow'])
plt.xlabel('EXCEL CONHECIMENTO')
plt.ylabel('Quantidade de Pessoas')
plt.title('Quantidade de Pessoas que tem conhecimento excel')
plt.xticks(rotation=0)
plt.show()
```
